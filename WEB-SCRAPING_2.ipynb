# Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and salary filter.

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import pandas as pd
```

```python
with open('ASSIGNMENT_naukri.csv', 'w') as file:
    writer = csv.writer(file, delimiter=',')
    writer.writerow(['Title', 'Location', 'Company', 'Experience Required\n'])
    
driver_naukri = webdriver.Chrome()
# Navigate to the Shine.com website
driver_naukri.get('https://www.naukri.com/')
time.sleep(1)


# Enter "Data Scientist" in the search field
designations_shine = driver_naukri.find_element(By.XPATH, '/html/body/div[1]/div[7]/div/div/div[1]/div/div/div/div[1]/div/input')
designations_shine.send_keys('Data Scientist')
time.sleep(1)


location = driver_naukri.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input')
location.send_keys('Bangalore')
time.sleep(1)

# Click on the search button
search = driver_naukri.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[6]')
search.click()
time.sleep(1)

salary=driver_naukri.find_element(By.XPATH, '/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[4]/div[2]/div[2]/label/p/span[1]')
salary.click()
time.sleep(1)

for k in range(5):
    # Extract job titles, locations, companies, and experience required
    titles = driver_naukri.find_elements(By.XPATH, '//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]')
    locations = driver_naukri.find_elements(By.XPATH, '//span[@class="locWdth"]')
    companies = driver_naukri.find_elements(By.XPATH, '//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]')
    experiences = driver_naukri.find_elements(By.XPATH, '//span[@class="expwdth"]')

    # Find the minimum length to avoid index out of range errors
    # Append job data to CSV
    with open('ASSIGNMENT_naukri.csv', 'a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file, delimiter=',')
        for i in range(20):  # Write only up to the minimum length
            writer.writerow([titles[i].text, locations[i].text, companies[i].text, experiences[i].text])

     
        next_button = driver_naukri.find_element(By.XPATH, '//a[@class="styles_btn-secondary__2AsIP"]')
        next_button.click()
        time.sleep(2)  # Allow some time for the next page to load

driver_naukri.quit()
```

```python
Naukri=pd.read_csv('ASSIGNMENT_naukri.csv')
Naukri
```

```python
import regex as re

title_naukri = []

for element in Naukri.Title:
    pattern=re.compile('\n')
    match = re.split(pattern, element)
    title_naukri.append(match[0])  # Get the first part before the newline

# job_location_shine will now contain the extracted job titles

Naukri['Title']=title_naukri
```

```python
company_naukri = []

for element in Naukri.Company:
    pattern = r'^[^\n]*\n([^\n]*)'
    match = re.search(pattern, element)
    if match:
        company_naukri.append(match.group(1))

Naukri['Company']=company_naukri
```

```python
for k in range(3):
    brand_sneakers=driver_sneakers.find_elements(By.XPATH,'//div[@class="syl9yP"]')
    product_Description_sneakers=driver_sneakers.find_elements(By.XPATH,'//div[@class="hCKiGj"]')
    price_sneakers=driver_sneakers.find_elements(By.XPATH,'//div[@class="Nx9bqj"]')


    with open('sneakers.csv', 'a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file, delimiter=',')
        for i in range(40):
            writer.writerow([brand_sneakers[i].text, product_Description_sneakers[i].text, price_sneakers[i].text])

        next_button = driver_sneakers.find_element(By.XPATH, '//a[@class="_9QVEpD"]')
        next_button.click()
        time.sleep(2)
driver_sneakers.quit() 
```

```python

title_laptop=driver_laptop.find_elements(By.XPATH,'//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]')
rating_laptop=driver_laptop.find_elements(By.XPATH,'//div[@class="a-row a-size-small"]')
price_laptop=driver_laptop.find_elements(By.XPATH,'//span[@class="a-price-whole"]')

with open('amazon_Laptop.csv', 'a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file, delimiter=',')
        for i in range(len(title_laptop)):
            writer.writerow([title_laptop[i].text, rating_laptop[i].text, price_laptop[i].text])
driver_laptop.quit()
```

# Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.

```python
with open('azquotes.csv','w') as file:
    write=csv.writer(file,delimiter=',')
    write.writerow(['Quote', 'Author' ,'Type_Of_Quotes'])
    
drivers_azquotes=webdriver.Chrome()
drivers_azquotes.get('https://www.azquotes.com/')

top_quotes=drivers_azquotes.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
top_quotes.click()
```

```python
for k in range(10):
    Quote=drivers_azquotes.find_elements(By.XPATH,'//a[@class="title"]')
    Author=drivers_azquotes.find_elements(By.XPATH,'//div[@class="author"]')
    Type_Of_Quotes=drivers_azquotes.find_elements(By.XPATH,'//div[@class="tags"]')

    with open('azquotes.csv','a',newline='',encoding='utf-8') as file:
        write=csv.writer(file,delimiter=',')
        for i in range(100):
            write.writerow([Quote[i].text,Author[i].text,Type_Of_Quotes[i].text])


        next_botton=drivers_azquotes.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a')
        next_botton.click()
drivers_azquotes.quit()
```
