```python
name=soup_1.find('h3',class_="ipc-title__text")
print(name.text)


rating=soup_1.find('span',class_="ipc-rating-star--rating")
print(rating.text)

YPR=soup_1.find('span',class_="sc-b189961a-8 hCbzGp dli-title-metadata-item")
print(YPR.text)
```

```python
#Extracting all the name from the given website
title_names=[]

for i in soup_1.find_all('h3',class_="ipc-title__text"):
    title_names.append(i.text)
print('-' * 127) 
print('Title Names')
print(title_names)
print('-' * 127)
#title name contains number as well Eg:'1. Ship of Theseus' and also we had unwanted data at the end, 
#so by the below code we can extract only the movies name's as we required.

name=[]

for i in title_names:
    p=re.compile('\A[0-9.]+')
    m=re.sub(p,"",i)
    name.append(m)
    
#Created a series 
Name=pd.Series(name[:-2])
print('-' * 127) 
print(Name)
print('-' * 127) 

#Below code is used to extracting all the Rating from the given website

title_rating=[]
for i in soup_1.find_all('span',class_="ipc-rating-star--rating"):
    title_rating.append(i.text)

print('-' * 127) 
print(title_rating)
print('-' * 127)
    

#Below code is used to extracting all the Year of release from the given website

title_year=[]

for i in soup_1.find_all('div',class_="sc-b189961a-7 btCcOY dli-title-metadata"):
    title_year.append(i.text)

print('-' * 127) 
print(title_year)# As the findall function find all the data which also contains year and time ,so further we will extract data as required 
print('-' * 127)

year_of_release=[]

for i in title_year:
    p=re.compile('[0-9][0-9][0-9][0-9]')
    m=re.findall(p,i)
    year_of_release.append(m)
    
print('-' * 127) 
print(year_of_release)
print('-' * 127)


IMDB=pd.DataFrame({'Name':Name,'Rating':title_rating,'Year_of_Release':year_of_release})
IMDB
```

```python
title_house=[]

for i in soup_2.find_all('a',class_="overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full"):
    title_house.append(i.text)


print(pd.DataFrame(title_house))
print('-' * 125) 
print('-' * 125) 
          
# Extracting the first three words

house=[]

for s in title_house:
    pattern = re.compile(r"^([\w+]+\s[\w+]+\s[\w+]+)")
    matche_title_house = re.search(pattern, s).group(1) 
    house.append(matche_title_house)

House=pd.DataFrame(house,columns=['Title'])
print(House)
print('-' * 125) 
print('-' * 125) 

title_location=[]

for i in soup_2.find_all('div',class_="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0.1p po:max-w-95"):
    title_location.append(i.text)
    
Loaction=pd.DataFrame(title_location,columns=['Location'])


print(Loaction)
print('-'*125)
print('-'*125)


title_area=[]

for i in title_house:
    p=re.compile(r'(\s\w+).+(\s\w+).+(\s\w+)')
    m=re.search(p,i).group(3)
    title_area.append(m)
    
Area=pd.DataFrame(title_area,columns=['Area'])
print(Area)
print('-'*125)
print('-'*125)


title_price=[]

for i in soup_2.find_all('div',class_="p-1.5p flex border-b border-b-solid border-cardbordercolor tp:py-1p tp:px-1.5p tp:border-b-0"):
    title_price.append(i.text)


price=pd.DataFrame(title_price)
price[0]=price[0].str.replace('₹', '')
print(price)
print('-'*125)
print('-'*125)

title_Price=[]

for i in price[0]:
    p1=re.compile(r'(([0-9].[0-9]\s)|(\d)+\s)[a-zA-Z]')
    m=re.search(p1,i)
    title_Price.append(m.group(0))



Price=pd.DataFrame(title_Price,columns=['Price'])

print(Price)
print('-'*125)
print('-'*125)

results_emi = []

pattern =re.compile( r"[\d,]+(?:\.\d+)?(?:\sLacs|\sCrores)?/MonthEstimated")

for text in title_price:
    matches = re.findall(pattern, text)
    for match in matches:
        results_emi.append(match)

        
EMI = pd.DataFrame(results_emi, columns=['Emi'])
print(EMI)
print('-'*125)
print('-'*125)
```

```python
NO_Broker=pd.concat([House,Loaction,Area,Price,EMI],axis=1)
NO_Broker
```

# Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/bestseller?sort=popular 

```python
url_4=requests.get('https://www.bewakoof.com/bestseller?sort=popular%20')
url_4
```

```python
soup_4=BeautifulSoup(url_4.content)
```

```python
name=soup_4.find('div',class_="d-flex")
print(name.text)
price=soup_4.find('div',class_="discountedPriceText clr-p-black false")
print(price.text)
img=soup_4.find('div',class_="productImg")
img
```

```python
title_name=[]

for i in soup_4.find_all('div',class_="productNaming bkf-ellipsis"):
    title_name.append(i.text)
    
Name=pd.DataFrame(title_name,columns=['Product'])
print(Name)
print('-'*125)
print('-'*125)

title_price=[]

for i in soup_4.find_all('div',class_="discountedPriceText clr-p-black false"):
    title_price.append(i.text)
    
Price=pd.DataFrame(title_price,columns=['Price'])
print(Price)
print('-'*125)
print('-'*125)


title_img = []

for i in soup_4.find_all('img',class_="productImgTag"):
    # Try to get 'data-src', fallback to 'src' if 'data-src' doesn't exist
    img_url = i.get('data-src') or i.get('src')
    if img_url:
        title_img.append(img_url)

Img=pd.DataFrame(title_img,columns=['Image'])
print(Img)
print('-'*125)
print('-'*125)
```

```python
bewakoof=pd.concat([Name,Price,Img],axis=1)
bewakoof
```
